#CSV file is not encoded in UTF-8, which is the default encoding used by pandas.read_csv() ,means that the file has special characters (like non-breaking spaces or smart quotes) that utf-8 can't interpret.

```python
df = pd.read_csv(file_path, encoding='ISO-8859-1')
# or
# df = pd.read_csv(file_path, encoding='cp1252')
```

```python
df.head() #shows the first few rows
```

```python
df.info() #summary of cols and dtypes
```

```python
df.describe() # basic stats of numerical cols
```

cleaning the extra spaces in csv columns if present.

```python
df.columns = df.columns.str.strip() # removing leading/trailing spaces
```

```python
print(df.columns.tolist()) # viewing the cleaned col names
```

```python
 #total sales ko hisab
print(f"Total Sales: ${df['Sales'].sum():.2f}")

#total qnty ko hisab
print(f"Total Quantity: {df['Quantity'].sum()}")

#avg dis ko hisab
print(f"Average Discount: {df['Discount'].mean():.2f}")

#total profit ko hisab
print(f"Total Profit: ${df['Profit'].sum():.2f}")
```

# Chekcing the profit by category

```python
# Group the data by 'Category' and add up the 'Profit' for each
profit_by_category = df.groupby('Category')['Profit'].sum()

# Create a bar chart to show profit for each category
profit_by_category.plot(kind='bar', title='Profit by Category', ylabel='Profit ($)', xlabel='Category')

# Adjust layout to make the plot fit well
plt.tight_layout()

# Show the plot
plt.show()
```

# Discount vs Profit scatter plot

```python
# Create a scatter plot with Discount on the X-axis and Profit on the Y-axis
sns.scatterplot(data=df, x='Discount', y='Profit')  # Shows how profit changes with different discount values

# Add a title to the plot
plt.title('Discount vs Profit')

# Show the scatter plot
plt.show()
```

# Creating each table one by one

## File with only customer info (customer.csv)

```python
#Imagine organizing a customer contact list and you only want unique customers

#picking only the cols realted to cust info
customer = df[['Customer ID','Customer Name','Segment','Region']]

# Remove duplicates (some customers appear in many orders, but we only need them once)
customer = customer.drop_duplicates()

# Savingcleaned data to a new CSV file
customer.to_csv('customer.csv', index=False)

```

```python
customer.head(20)
```

# product.csv

```python
#list every item you sell just once

# Picking only product related columns
product = df[['Product ID', 'Product Name', 'Category', 'Sub-Category']]

# Removed repeated rows as the same product might be sold multiple times
product = product.drop_duplicates()

# Saved it to CSV
product.to_csv('product.csv', index=False)
```

```python
product.head()
```

# Order.csv

```python
#Each order is like a receipt. Even if it has many products the order ID stays the same.

# Picking order details (linking it to Customer ID so we know who ordered)
order = df[['Order ID', 'Order Date', 'Ship Date', 'Customer ID']]

# Removed repeated orders
order = order.drop_duplicates()

# Saved to CSV
order.to_csv('order.csv', index=False)
```

```python
order.head()
```

# Shipmode.csv

```python
# just shows all the ways people can get their stuff delivered

# Get all unique shipping methods
shipmode = df[['Ship Mode']].drop_duplicates()

# Save to CSV
shipmode.to_csv('shipmode.csv', index=False)
```

```python
shipmode.head()
```

# Sales.csv

```python
# This is the table that connects everything together

sales = df[['Order ID', 'Product ID', 'Customer ID', 'Ship Mode', 'Sales', 'Quantity', 'Discount', 'Profit']]

# Save to CSV
sales.to_csv('sales.csv', index=False)
```
