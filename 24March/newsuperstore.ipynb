```python
import pandas as pd
```

```python
file_path = r"C:\Users\ACER\Desktop\JAVASCRIPT BASICS\Sample - Superstore.csv"
```

```python
try:
    df = pd.read_csv(file_path, encoding="utf-8")  # Try UTF-8 first
except UnicodeDecodeError:
    print("UTF-8 failed. Trying ISO-8859-1...")
    df = pd.read_csv(file_path, encoding="ISO-8859-1")  # Fallback encoding
```

```python
df.info()
```

```python
df.head()
```

```python
order_columns = ["Order ID", "Order Date", "Ship Date", "Ship Mode", "Customer ID", "Segment", "Country", "Region"]
order_df = df[order_columns]
```

```python

order_df = order_df.drop_duplicates(subset=["Order ID"])
```

```python

order_csv_path = "orders.csv"
order_df.to_csv(order_csv_path, index=False)

print(f"Order-related data saved successfully as '{order_csv_path}'.")
```
